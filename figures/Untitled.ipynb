```python
penalties = (1.75, 2.0, 3.0)
```

```python
dfs = {
    p: pd.read_csv(f"../data/Qfile_short_bf_{p}_{p}.csv")
    for p in penalties
}
```

```python
for p in penalties:
    dfs[p]["Penalty"] = f"$p_{{pair}} = p_{{sum}} = {p}$"
```

```python
total_df = pd.concat(list(dfs.values()))
```

```python
total_df.head()
```

```python
f = sns.histplot(df.truncate(after=5000), hue="feasibility", x="energy", palette=["red", "green"])
f.set_yscale("log")
```

```python
g = sns.FacetGrid(total_df, col="Penalty", sharex=False, legend_out=True)
g.map_dataframe(sns.histplot, x="energy", hue="feasibility", alpha=0.5, palette=["red", "green"])

import matplotlib.patches as mpatches
red_patch = mpatches.Patch(color='red', alpha=0.5, label='Infeasible')
green_patch = mpatches.Patch(color='green', alpha=0.5, label='Feasible')

for ax in g.axes[0]:
    ax.set_yscale("log")
    ax.set_xlabel("Energy")
    ax.legend(handles=[green_patch, red_patch], ncols=2)
#g.axes[0][2].legend()
g.set_titles(col_template="{col_name}")


plt.savefig("./railway_histograms_bf.pdf")
```
